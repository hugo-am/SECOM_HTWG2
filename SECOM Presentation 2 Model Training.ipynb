```python
import pandas as pd
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree
```

```python
#Generate an empty list to put column names in
column_list=[]
```

```python
#Reading csv file 1
labels=pd.read_csv('secom_labels.data',header=None,sep=' ',names=['status','timestamp'])
```

```python
data=pd.read_csv('secom.data',header=None,sep=' ')
```

```python
#Naming columns 0-590
for x in range(591):
    column_list.append(f'col_{x}')
```

```python
#Remove the first column name as it starts with 0
column_list.remove('col_0')
```

```python
#Reading csv file 2
data=pd.read_csv('secom.data',header=None,sep=' ',names=column_list)
```

```python
#The secom.names file indicates that both .data files have the same order, therefore we can merge using the index
df=data.merge(labels,left_index=True,right_index=True)
```

```python
df['status']=df['status'].replace(-1,0)
```

```python
col_list=['col_132','col_58','col_38','col_39','col_122','col_134','col_583','col_124','col_125','col_57','col_110','col_56','col_112','col_120','col_117','col_21','col_54','col_55','col_18','col_548','col_114','col_12','col_88','col_3','col_119','col_51','col_19','col_40','col_71','col_87']
tgt=['status']
```

```python
#30 columns with the most information "covariance"

df_selection=df[col_list+tgt].copy()
```

```python
df_selection['col_119'].fillna(df_selection['col_119'].median())
```

```python
for col in col_list:
    print(col)
    df_selection[col]=df_selection[col].fillna(df_selection[col].median())
    
```

```python
df_selection
```

```python
#train split
Xt,Xv,yt,yv=train_test_split(df_selection[col_list],df_selection[tgt],train_size=0.7)
```

```python
#scaling
mm=MinMaxScaler()
Xs=pd.DataFrame(mm.fit_transform(Xt[col_list]),columns=col_list)
```

```python
#decision tree
tr=tree.DecisionTreeClassifier()
tr=tr.fit(Xt,yt)
```

```python
tr.predict(Xv).sum()
```

```python
tr.score(X=Xv,y=yv)
```
